```python
university_rating_num = university.groupby('institution').count()['score']#Number of scores for particular institution,we will be classifying based on gre score and rating
university_rating_num = university_rating_num.rename('University Education Quality').reset_index()#rename 
university_rating_num
```

```python
university_rating_avg = university.groupby('institution').mean()['score'].reset_index()
university_rating_avg.rename(columns={'score': 'University Education Quality Average'}, inplace=True)
university_rating_avg
```

```python
university_popularity_df = university_rating_num.merge(university_rating_avg, on="institution")
university_popularity_df
```

```python
popular_university_df=university_popularity_df[university_popularity_df['University Education Quality Average'] >= 80]#selecting top universities
```

```python
def recommendation(institution_name, top_n=10):
    try:
        institution_name = institution_name.strip().lower()  # Preprocessing input

        # Find similar institution names in lowercase
        similar_names = university['institution'].str.strip().str.lower().str.contains(institution_name)

        # Check for similarity
  # Check for similarity
        if similar_names.any():
            index = similar_names[similar_names].index[0] 
            similarity_scores = similarity_score[index]
            similar_indices = similarity_scores.argsort()[::-1][1:top_n+1]
            
            print(f"Universities similar to {university.loc[index, 'institution']}:")
            for i in similar_indices:
                print(university.loc[university.index[i], 'institution'])
        else:
            print(f"No similar universities found for {institution_name.capitalize()}")
    except IndexError:
        print(f"No similar universities found for {institution_name.capitalize()}")
```
